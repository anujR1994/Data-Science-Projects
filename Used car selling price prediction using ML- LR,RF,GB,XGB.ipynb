#### Encoding the Categorical Columns

For encoding the categorical variables, we will consider what variables are nominal and what are ordinal in nature:

- Nominal Variables:
    Nominal variables are categorical variables where the categories do not have a natural order or ranking.

    The categories are distinct and unrelated, and there is no inherent order or hierarchy among them.
       Examples:
            Colors (e.g., red, blue, green)
            Countries (e.g., USA, Canada, Japan)
            Types of fruits (e.g., apple, banana, orange)
            
    **For nominal variables we prefer one hot encoding.**

- Ordinal Variables:
    Ordinal variables are categorical variables where the categories have a natural order or ranking.
    
    The categories represent ordered levels or ranks, where one category is considered "higher" or "lower" than another.
    The differences between the categories are meaningful, but the exact magnitude of the differences may not be specified.

    **For ordinal variables we prefer label encoding**
    
        Examples:
            Education level (e.g., high school, bachelor's, master's, Ph.D.)
            Likert scale responses (e.g., strongly disagree, disagree, neutral, agree, strongly agree)
            Socioeconomic status (e.g., low income, middle income, high income)

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd

# Initialise the label encoder before applying it to ordinal variable 'Owner'
le = LabelEncoder()

# Applying label encoder to 'Owner' column
data['Owner_encoded'] = le.fit_transform(data['Owner'])

# Drop the original 'Owner' column
data.drop(columns=['Owner','Name'], inplace=True)

# Encode the 'Sold' column
data['Sold_encoded'] = le.fit_transform(data['Sold'])

# Drop the original 'Sold' column
data.drop(columns=['Sold'], inplace=True)

# Get dummy variables for categorical columns and drop original columns
df_encoded = pd.get_dummies(data, columns=['Region', 'State or Province', 'Fuel', 'Seller_type', 'Transmission'], drop_first=True)

# Combine the encoded columns with the original DataFrame
data = pd.concat([data.drop(columns=['Region', 'State or Province', 'Fuel', 'Seller_type', 'Transmission']), df_encoded], axis=1)
```

## Regularization Techniques in Linear Regression

### Ridge Regression
Ridge regression addresses overfitting by penalizing the size of coefficients with L2 regularization. It's particularly useful when dealing with multicollinearity among predictors.

### Lasso Regression
Lasso regression estimates sparse coefficients by imposing L1 regularization. It automatically performs feature selection by driving some coefficients to zero, making it effective for high-dimensional data.

### ElasticNet Regression
ElasticNet combines L1 and L2 regularization, offering a balance between Ridge and Lasso. It's useful when there are multiple correlated predictors, as it tends to select groups of correlated predictors together.

Regularization techniques are essential for improving model generalization and robustness in linear regression.

Read more:
